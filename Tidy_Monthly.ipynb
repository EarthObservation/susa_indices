# 1. Monthly S2-indices for 2017 (DONE)

- Calculate monthly mean
- Interpolate monthly (linear)
- Join all indices into a single file

```python
import pandas as pd
```

```python
%%time

suff = "savi"
        
src_pth = f".\\results_extracted\\ZV2017_extracted_{suff}.csv"

df0 = pd.read_csv(src_pth)
df0.tail(3)
```

```python
%%time

suffs = ["evi2", "ndvi", "ndwi", "savi"]

# Merge all 4 indices into a single file
new_dfs = []
for suff in suffs:
    df = pd.read_csv(f".\\results_extracted\\ZV2017_extracted_{suff}.csv")
    df = df.drop(columns=["id_17"])
        
    for month in range(4, 11):
        # Find columns containing current band and month
        tmp_cols = [col for col in df.columns if f"{suff}_{month:02}" in col]

        # Append new column name to the list of mean columns
        new_col = f"{suff}_{month:02}"

        # Calculate nanmean
        df[new_col] = df[tmp_cols].mean(axis=1)
        df = df.drop(columns=tmp_cols)
        
    # Interpolate missing values
    df = df.interpolate(limit_direction="both", axis=1)
    
    # Append to list for concating
    new_dfs.append(df)
    
df = pd.concat(new_dfs, axis=1)

# Add the "id_17" index column
df = df.reset_index()
df = df.rename(columns={"index": "id_17"})

df.tail(2)
```

```python
df.to_csv(f".\\results_monthly\\ZV2017_s2-indices_monthly.csv", index=False)
```
