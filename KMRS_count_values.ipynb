# Crop-type (SIFRA_KMRS) instances through years

This notebook compares how many instances of each crop-type there is for years 2018, 2019 and 2020.
I have only processed (finding nearest three neighbours) those crop-types that had 1000 or more instances in each year.

```python
import geopandas as gpd
import pandas as pd
```

```python
# Import shape files
near_17 = pd.read_csv(".\\results_monthly/\\ZV2017_s2-indices_monthly.csv")
near_18 = pd.read_csv(".\\results_monthly/ZV2017-2018_s2-indices_monthly.csv")
near_19 = pd.read_csv(".\\results_monthly/ZV2017-2019_s2-indices_monthly.csv")
```

```python
# Import shape files
clean_17 = gpd.read_file(".\\poljine_SHP\\ZV2017_d96tm_clean_ids.shp")
clean_18 = gpd.read_file(".\\poljine_SHP\\ZV2018_d96tm_clean_ids.shp")
clean_19 = gpd.read_file(".\\poljine_SHP\\KMRS2019_d96tm_clean_ids.shp")
```

```python
# Value counts for all 3

vc_17 = clean_17.SIFRA_KMRS.value_counts().rename("ZV2017").convert_dtypes()
vc_18 = clean_18.SIFRA_KMRS.value_counts().rename("ZV2018").convert_dtypes()
vc_19 = clean_19.SIFRA_KMRS.value_counts().rename("KMRS2019").convert_dtypes()
```

```python
# Read sifrant and map names to KMRS_numbers
excel = pd.read_csv(".\\excel\\Sifrant-KMRS-2020.csv").convert_dtypes()
excel = excel.rename(columns={"KMRS": "SIFRA_KMRS"})
excel["SIFRA_KMRS"] = excel["SIFRA_KMRS"].astype("Int64")
excel.head(3)
```

```python
pd.set_option('display.max_rows', 500)
final = kmrs.merge(excel, how="left")

final.to_csv(".\\excel\\Sifrant-KMRS-2020_primerjava.csv")
```
