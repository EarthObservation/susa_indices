# Prepare clean 2017 file

```python
# Import the starting "clean" version of the 2017 file
org_2017 = ".\\poljine_SHP\\ZV2017_d96tm_clean.shp"
# Import the resulting HAND dataset for the year 2017
hand_2017 = ".\\poljine_SHP\\ZV2017_d96tm_pc075_hand.csv"
```

```python
# Read files
df_clean17 = gpd.read_file(org_2017)
df_hand17 = pd.read_csv(hand_2017)
```

```python
df_clean17.tail()
```

```python
df_hand17.tail()
```

The `ZV2017_d96tm_clean.shp` file has extra lines that have not been included in the final table. These correspond to polygons hat are narrow and were dropped during the "buffernig" step of prreparing the shapefile.

```python
# This will filter the "clean" file to include only relevant lines

shp_2017 = ".\\poljine_SHP\\ZV2017_d96tm_pc075_hand.shp"
df_2017 = gpd.read_file(shp_2017)
df_2017.tail()
```

Drop columns from the source file (this shape was generated from a HAND dataset, that was created earlier in the project - I have never saved the "clean" version of the file, therfore I had to recreate it from one of the tables)

```python
df_2017 = df_2017.drop(columns=["pix_count", "hand_mean", "hand_std"])
df_2017.tail()
```

```python
# Add unique index
df_2017 = df_2017.reset_index()
df_2017 = df_2017.rename(columns={"index": "id_17"})
df_2017
```

```python
# Save clean
df_2017.to_file(".\\poljine_SHP\\ZV2017_d96tm_clean_ids.shp")
```

This is the final version of a clean ZV2017 shapefile. It includes:
- 664137 lines
- unique index called `id_17`
